```python
import seaborn as sns 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
```

```python
path = os.path.dirname(os.path.abspath('__file__'))+'\\Prepared_tables\\'
path_saving = os.path.dirname(os.path.abspath('__file__'))+'\\raw_figures'
df = pd.read_csv(path + 'Prep_Protein Groups_cut0_notimuted.csv', low_memory=False)

df['SubMitoLocalization'] = [df['SubMitoLocalization'][i] if str(df['SubMitoLocalization'][i]) !='nan' else '' for i in range(len(df))]
df['IMPI_new'] = [df['IMPI_new'][i] if str(df['IMPI_new'][i]) !='nan' else '' for i in range(len(df))]
```

```python
cols = [i for i in range(len(list(df.columns)))  if ('Norm_Intensity ' in list(df.columns)[i]) & ('Zscore' not in (df.columns)[i]) and ('wt' not in (df.columns)[i])]
rows = [i  for i in range(len(df)) if ((df['SubMitoLocalization'][i]!='') or
                                                                         ('Verified mitochondrial' in df['IMPI_new'][i])==True)] #or 
                                                                         #('Predicted mitochondrial' in df['IMPI_new'][i])==True)]
df_mito = df.iloc[rows,:]
df_heatmap = df_mito.iloc[:,cols].corr(method='pearson',min_periods = 1)
bar_color = sns.color_palette("coolwarm", as_cmap=True)
#bar_color = sns.diverging_palette(250, 30, l=65, center="dark", as_cmap=True)
Cluster_color = ['#66c2a5',"#fc8d62","#8da0cb","#e78ac3","#a6d854","#ffd92f","#e5c494"] #colorfull 
#Cluster_color = ['#252525',"#525252","#737373","#969696","#bdbdbd","#d9d9d9","#f0f0f0"] #blacks
row_colors = [Cluster_color[0],
                        Cluster_color[0],
                        Cluster_color[0],
                        Cluster_color[0],
                        Cluster_color[0],
                        Cluster_color[0],
                        Cluster_color[1],
                        Cluster_color[1],
                        Cluster_color[1],
                        Cluster_color[1],
                        Cluster_color[1],
                        Cluster_color[1],
                        Cluster_color[2],
                        Cluster_color[2],
                        Cluster_color[2],
                        Cluster_color[2],
                        Cluster_color[2],
                        Cluster_color[2],
                        Cluster_color[3],
                        Cluster_color[3],
                        Cluster_color[3],
                        Cluster_color[3],
                        Cluster_color[3],
                        Cluster_color[3],
                        Cluster_color[4],
                        Cluster_color[4],
                        Cluster_color[4],
                        Cluster_color[4],
                        Cluster_color[4],
                        Cluster_color[4],
                        Cluster_color[5],
                        Cluster_color[5],
                        Cluster_color[5],
                        Cluster_color[5],
                        Cluster_color[5],
                        Cluster_color[5],
                        Cluster_color[6],
                        Cluster_color[6],
                        Cluster_color[6],
                        Cluster_color[6],
                        Cluster_color[6],
                        Cluster_color[6]
                       ]
g = sns.clustermap(df_heatmap, cmap =bar_color, linewidths = 0.0, annot = False, 
                   vmin=0.6, vmax=1, 
                   #row_colors =row_colors, col_colors =row_colors,
                   square=True,cbar_pos = (1,0.0175,0.05,0.25),
                   yticklabels=True,xticklabels=False,
                   metric="correlation") 
plt.savefig(path_saving+'\\Figure_2\\Prot_correlation_Mito.pdf', bbox_inches = "tight")
```

```python
df_heatmap.to_csv((path_saving+ '\\Figure_2\\Prot_correlation_Mito_heatmap.csv'))
```
