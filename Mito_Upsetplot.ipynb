```python
import pandas as pd
from upsetplot import plot
from matplotlib import pyplot as plt
import sys
import os
import numpy as np
```

<h1> UpSetPlot: Mitochondrial Proteins

```python
path = os.path.dirname(os.path.abspath('__file__'))+'\\Prepared_tables\\'
path_saving = os.path.dirname(os.path.abspath('__file__'))+'\\raw_figures\\UpSet\\'

df_prot = pd.read_csv((path+ 'Prep_Protein Groups_mito.csv'), low_memory= False)

lower_lim = 0
group_valid_value = 0.01


B = (df_prot["Valid values_B_"] >= group_valid_value)
BAT = (df_prot["Valid values_BAT_"] >= group_valid_value)
H = (df_prot["Valid values_H_"] >= group_valid_value)
K = (df_prot["Valid values_K_"] >= group_valid_value)
L = (df_prot["Valid values_L_"] >= group_valid_value)
S = (df_prot["Valid values_S_"] >= group_valid_value)
SKM = (df_prot["Valid values_SKM_"] >= group_valid_value)

new_lst = [list(x) for x in zip(B, BAT, H,K, L,S, SKM)]
lists = [list(item) for item in set(tuple(row) for row in new_lst)]
value= []
for i in range(len(lists)):
    value.append(new_lst.count(lists[i]))

lists_index_filter = [i for i, e in enumerate(value) if e > lower_lim]
value_filter = [ i for i in value if i > lower_lim ]

index = pd.MultiIndex.from_tuples(lists, names=['Brain', 'BAT','Heart', 'Kidney'  , 'Liver','Spleen', 'SKM'])

s = pd.Series(value_filter, index=index[lists_index_filter])


plot(s, show_counts= True, sort_by= 'cardinality', totals_plot_elements= 5,intersection_plot_elements=7)
current_figure = plt.gcf()
current_figure.axes[3].set_ylabel('Intersection size',size='xx-large', family='Arial')
current_figure.axes[1].set_yticklabels(current_figure.axes[1].get_yticklabels(),size='x-large', family='Arial')
current_figure.axes[3].set_xbound(-0.5,9.5)
#current_figure.axes[3].texts= current_figure.axes[3].texts[:10]
current_figure.set_size_inches(15,10)


current_figure.axes[2].change_geometry(3,6,9)
current_figure.axes[3].change_geometry(3,2,2)
current_figure.axes[1].change_geometry(3,2,4)
current_figure.axes[0].change_geometry(3,2,4)

#current_figure.savefig((path_saving + r'Upsetplot_prot_1Valid.pdf'))

exclusive_prot = len([i for i in range(len(new_lst)) if sum(new_lst[i]) == 1])
two_tissue_prot = len([i for i in range(len(new_lst)) if sum(new_lst[i]) == 2])
common_prot = len([i for i in range(len(new_lst)) if sum(new_lst[i]) == 7])
all_prot = len(new_lst)
```

<h1> Barplot: Protein+STYsites

```python
fig, ax = plt.subplots(figsize=(8, 4))
x = np.arange(2)
bar_width = 0.2

temp = "ID_all"
#res = [val[temp] for key, val in site_dict_sorted.items() if temp in val]
b1 = ax.bar(x, [all_prot,all_phos],width=bar_width, label='All identifications', color = '#737373', ec = 'black')
temp = "ID_ClassI"
#res = [val[temp] for key, val in site_dict_sorted.items() if temp in val]
b2 = ax.bar(x+bar_width, [common_prot,common_phos],width=bar_width, label='Common to all tissues', color = '#969696', ec = 'black')
temp = "ID_Mito_ClassI"
#res = [val[temp] for key, val in site_dict_sorted.items() if temp in val]
b3 = ax.bar(x+2*bar_width,[exclusive_prot,exclusive_phos] ,width=bar_width, label='Tissue exclusive', color = '#bdbdbd', ec = 'black')


ax.set_xticks(x + bar_width)
ax.set_xticklabels(['Proteins','STY sites'], size = 16)

ax.legend(fontsize = 'large', facecolor='white', framealpha = 1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('black')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='black')
#ax.xaxis.grid(False)

#ax.set_xlabel('Tissues', labelpad=15,size = 18,family='Arial')
ax.set_ylabel('# Identifications', labelpad=15,size = 18,family='Arial')
ax.set_title('', pad=15)
ax.set_facecolor("white")
 
fig.tight_layout()
path_saving = os.path.dirname(os.path.abspath('__file__'))+'\\raw_figures\\BarPlots\\'
#fig.savefig((path_saving + r'Protein_STYsites_1Valid.pdf'))
```
