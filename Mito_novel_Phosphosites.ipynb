```python
import pandas as pd
import numpy as np
import sys
import os
from collections import Counter
import matplotlib.pyplot as plt
```

```python
path_databases = os.path.dirname(os.path.abspath('__file__'))
df_known_sites = pd.read_csv((path_databases+'\\known_PSP_Phosphosites_mouse.csv'),low_memory=False)

path = os.path.dirname(os.path.abspath('__file__'))+'\\Prepared_tables\\'
df_sites = pd.read_csv((path+'Prep_STYsites_mito.csv'),low_memory=False)
```

```python
df_sites['Proteins'] = [df_sites['Proteins'][i].split(';') for i in range(len(df_sites))]
df_sites['Positions within proteins'] = [df_sites['Positions within proteins'][i].split(';') for i in range(len(df_sites))]
res = [[i + '_'+ j for i, j in zip(df_sites['Proteins'][x], df_sites['Positions within proteins'][x])] for  x in range(len(df_sites))]


not_in_PSP = []
for x in range(len(res)): 
    a = [res[x][i] in list(df_known_sites['sites']) for i in range(len(res[x]))]
    if True not in a:
        not_in_PSP.append(x) 

sites_not_in_PSP =  len(not_in_PSP)        
print(str(sites_not_in_PSP)+ ' novel phosphosites')     
```
