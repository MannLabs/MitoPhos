```python
import sys
import os
from pyteomics import fasta
import pandas as pd
import alphamap
from alphamap.importing import import_data
from alphamap.preprocessing import format_input_data
```

```python
path = os.path.dirname(os.path.abspath('__file__'))+'\\MQ_Output\\'
full_proteome_data = import_data(path+'evidence.txt')

path_fasta = os.path.dirname(os.path.abspath('__file__'))
full_mouse_fasta = fasta.IndexedUniProt(path_fasta+'\\20201019_Mus musculus_10090_UP000000589.fasta')

formatted_proteome_data = format_input_data(df=full_proteome_data, fasta = full_mouse_fasta, modification_exp = r'\[.*?\]')

path_saving = os.path.dirname(os.path.abspath('__file__'))+'\\Prepared_tables\\'
formatted_proteome_data.to_csv((path_saving+ 'formatted_proteome_data.csv'), index= False)
```
